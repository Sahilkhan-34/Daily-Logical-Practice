```python
# Question 19
factor = []
number = 95
for i in range(1,number//2):
    if number % i == 0:
        factor.append(i)

print(f'Factor of {number} is {factor}')
        
```

```python
# Question 20 Strong Number
number = 145
def factorial(n):
    if n > 1:
        return n * factorial(n-1)
    else:
        return n
num_list = []
for i in str(number):
    num = factorial(int(i))
    num_list.append(num)
    
if sum(num_list) == number:
    print(f'{number} is Strong number')
else:
    print(f'{number} is Not Strong number')
```

```python
# Question 21
number = 495
factor = []
for i in range(1,number//2):
    if number % i == 0:
        factor.append(i)
flag = 1
for i in factor:
    flag = flag * i
if number == flag:
    print(f'{number} is isPerfect Number')
else:
    print(f'{number} is not isPerfect Number')
```

```python
# Question 22 Perfect Square
import math
number = 144
for i in range(1,13):
    if i < math.sqrt(number+1):
        if i*i == number:
            print(f'{i} is perfect square of {number}')
            break
    else:
        print(f'{number} has no perfect square')
        break
```

```python
# Question 23 Automorphic
number = 25
squar = number*number
if number < 10:
    if squar%10 == number:
        print(f'{number} is Automorphic')
else:
    if squar%100 == number:
        print(f'{number} is Automorphic')
    else:
        print(f'{number} is not Automorphic')
```

```python
# Question 24
number = 153
list_num = [int(digit) for digit in str(number)]
summ = sum(list_num)
if number % summ == 0:
    print(f'{number} is Harshad Number')
else:
    print(f'{number} is not a Harshad number')
```
