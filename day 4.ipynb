```python
# Question 25 abundent number
number = 54
summ = 0
for i in range(1,number+1//2):
    if number % i == 0:
        summ  += i
if number < summ:
    print(f'{number} is abundent number')
else:
    print(f'{number} is not abundent number')     
```

```python
# Question 26
# number1 = 30
# number2 = 140
def devisor(number):
    summ = 0
    for i in range(1,number+1 // 2):
        if number % i == 0:
            summ += i
    return summ

number1 = int(input("Enter the First Number: "))
number2 = int(input("Enter the Second Number: "))
a = devisor(number1)/number1
b = devisor(number2)/number2
if a == b:
    print(f'{number1} and {number2} is friendly pair')
else:
    print(f'{number1} and {number2} is not friendly pair')
```

```python
# Question 26 HCF of 2 Numbers
number1 = int(input("Enter the First Number: "))
number2 = int(input("Enter the Second Number: "))
def hcf(number1, number2):
    while number2:
        temp = number1
        number1 = number2
        number2 = temp % number2
        
    return number1


print(f'HCF of {number1} and {number2} is {hcf(number1,number2)}')
```
