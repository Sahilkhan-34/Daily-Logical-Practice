```python
# Question 15
def check_Armstrong_number(number):
    cube_num = 0
    for i in str(number):
        cube_num = cube_num + int(i)**3
    if cube_num == number:
        return number

arm_num = []
for i in range(100,501):
    num = check_Armstrong_number(i)
    if num:
        arm_num.append(num)
print(arm_num)
```

```python
# fabonacci
def fabonacci(n):
    fabonacci_series = [0,1]
    while fabonacci_series[-1] < n:
        fabonacci_series.append(fabonacci_series[-1] + fabonacci_series[-2])
    return fabonacci_series 
        
print(fabonacci(200))
```

```python
# factorial 5!

def factorial(num):
    if num > 1:
        return num * factorial(num-1)
    else:
        return num
    
print(factorial(5))
```
